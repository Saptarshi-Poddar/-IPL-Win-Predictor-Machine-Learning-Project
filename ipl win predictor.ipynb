## Feature Engineering

```python
team =[ 'Chennai Super Kings', 'Delhi Capitals', 'Kolkata Knight Riders',
        'Mumbai Indians', 'Rajasthan Royals', 'Sunrisers Hyderabad',
        'Gujarat Titans', 'Lucknow Super Giants','Punjab Kings', 'Royal Challengers Bangaluru'] # List of current teams playing in  IPL
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team1'] = match_df['team1'].str.replace('Royal Challengers Bangalore', 'Royal Challengers Bangaluru')
match_df['team2'] = match_df['team2'].str.replace('Royal Challengers Bangalore', 'Royal Challengers Bangaluru')
match_df['team1'] = match_df['team1'].str.replace('Kings XI Punjab', 'Punjab Kings')
match_df['team2'] = match_df['team2'].str.replace('Kings XI Punjab', 'Punjab Kings')  
match_df['team1'] = match_df['team1'].str.replace('Gujarat Lions', 'Gujarat Titans') # Gujarat Lions is not playing in IPL anymore
```

```python
match_df
```

```python
match_df.merge(deliveries,on = 'match_id')   # Merging the match_df with deliveries 
```

```python
deliveries_df = match_df.merge(deliveries,on = 'match_id')
```

```python
deliveries_df = deliveries_df[deliveries_df ['inning'] == 2]   # Displaying the rows of the deliveries_df during 2nd innings as we want to find the probability during chase only
```

```python
deliveries_df.shape
```

```python
# Calculate cumulative sum of runs for each match
cumulative_runs = deliveries_df.groupby('match_id')['total_runs'].cumsum() 
cumulative_runs
```

```python
deliveries_df['current_score'] = cumulative_runs
```

```python
deliveries_df
```

```python
deliveries_df['target'] =deliveries_df['Total_Runs_Scoredin1st_innings'] + 1
```

```python
deliveries_df['runs_left'] = deliveries_df['target'] - deliveries_df['current_score']  # Calculating the runs left to chase
```

```python
deliveries_df
```

```python
deliveries_df['ball_left'] = 120 - (deliveries_df['over']*6 + deliveries_df['ball'])  # Calculating the balls left to play 
```

```python
deliveries_df
```

```python
wickets = deliveries_df.groupby('match_id')['is_wicket'].cumsum()
wickets
```

```python
deliveries_df['wicket_left'] = 10 - wickets  # Calculating the wickets left to fall
```

```python
deliveries_df
```

```python
# current runrate (crr) = current score / overs played
deliveries_df['current_runrate'] = deliveries_df['current_score'] / (deliveries_df['over'] + deliveries_df['ball']/6)  # Calculating the current run rate
```

```python
deliveries_df
```

```python
#required runrate (rrr) = (runs left *6) / balls left 
deliveries_df['required_runrate'] = (deliveries_df['runs_left'] * 6) / deliveries_df['ball_left']  # Calculating the required run rate
```

```python
deliveries_df
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
deliveries_df.apply(result, axis=1)
```

```python
deliveries_df['result'] = deliveries_df.apply(result, axis=1)  # Applying the result function to each row to determine if the batting team won or not
```

```python
deliveries_df
```

## Model Building
### Training machine learning models and evaluating performance

```python
X= final_df.drop('result', axis=1)  # Features for the model
y = final_df['result']  # Target variable
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1)  # Splitting the dataset into training and testing sets with a test size of 20%
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([('encoder', OneHotEncoder(sparse_output =False), ['batting_team', 'bowling_team','city'])], remainder='passthrough')  # One-hot encoding the categorical features

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[   # Creating a pipeline to streamline the preprocessing and model training
    ('step1', ct),  # Adding the column transformer to the pipeline
    ('step2', LogisticRegression(solver = 'liblinear' ))  # Adding the logistic regression model to the pipeline
])
```

```python
X_train.describe() # Describing the training features to get an overview of the data
```

```python
X_train.describe()  # Describing the training features to get an overview of the data
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)  # Making predictions on the test set using the fitted pipeline
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))  # Calculating the accuracy of the model
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))  # Displaying the confusion matrix to evaluate the model's performance
print("Classification Report:\n", classification_report(y_test, y_pred))  # Displaying the classification report to get detailed metrics like precision, recall, and F1-score 

```
